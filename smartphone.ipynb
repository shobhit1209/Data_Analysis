```
def camera_extractor(text):
  if "Quad" in text:
    return 4
  elif "Triple" in text:
    return 3
  elif "Dual" in text:
    return 2
  elif "0" in text:
    return 0
  else:
    return 1
```

```
num_rear_cameras=df["camera"].str.strip().str.split("&").str.get(0).apply(camera_extractor)
```

```
num_front_cameras=df["camera"].str.strip().str.split("&").str.get(1).fillna("0").apply(camera_extractor)
```

```
def has_card(text):
  if "Not" in text:
    return "No"
  elif "Hybrid" in text:
    return "Yes (Hybrid)"
  else:
    return "Yes"
```

```
has_memory_card=df["card"].apply(has_card)
```

```
df.insert(31,"has_memory_card",has_memory_card)
```

```
def extended_memory (text):
  if "Not" in text:
    return np.nan
  elif "upto" in text:
    x=text.split("upto")
    return x[-1]
  else:
    return "Not Specified"
```

```
extended_memory=df["card"].apply(extended_memory)
```
